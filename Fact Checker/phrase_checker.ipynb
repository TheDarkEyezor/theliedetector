```python
# phrase analysis.
from duckduckgo_search import DDGS
import numpy as np

blacklist = {"wikipedia"}

def get_context(phrase: str):
  context = ""
  results = DDGS().text(phrase, max_results=5, safesearch="on", backend="lite")
  for result in results:
    if any([x in result["href"] for x in blacklist]):
      print("skipping", result["href"])
    else:
      context += f"{result["title"]}: {result["body"]}"
  return context

get_context("bratwuurst")
```
